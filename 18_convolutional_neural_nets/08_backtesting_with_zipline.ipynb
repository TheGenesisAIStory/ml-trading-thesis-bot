# Vectorized CNN Signal Backtest

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from cnn_utils import *

sns.set_theme(style='whitegrid', context='notebook')
pd.set_option('display.float_format', lambda x: f'{x:,.4f}')
OUTPUT = ensure_output_dir()
```

```python
X, y = trading_image_features()
model, accuracy = train_trading_image_model()
scores = pd.Series(model.predict_proba(X)[:, 1], name='score')
returns = pd.Series((scores - 0.5) * (y * 2 - 1) / 50, name='strategy_return')
backtest = pd.DataFrame({'strategy_return': returns, 'equity_curve': (1 + returns).cumprod()})
backtest.to_parquet(OUTPUT / 'cnn_signal_backtest.parquet', index=False)
backtest.tail()
```
