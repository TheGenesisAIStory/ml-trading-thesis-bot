# Coefficient Diagnostics

Modernized for Python 3.10+ without legacy optional dependencies.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from linear_model_utils import regression_data, classification_data, panel_factor_data, ols_numpy, rank_ic
OUTPUT = FOLDER.parent / 'data' / 'linear_models'
OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
x, y = regression_data(seed=5)
coef = ols_numpy(x, y)
design = np.column_stack([np.ones(len(x)), x.to_numpy()])
resid = y.to_numpy() - design @ coef.to_numpy()
sigma2 = (resid @ resid) / (len(y) - design.shape[1])
se = np.sqrt(np.diag(sigma2 * np.linalg.pinv(design.T @ design)))
diag = pd.DataFrame({'coef': coef, 'std_error': se}, index=coef.index)
diag['t_stat'] = diag['coef'] / diag['std_error']
diag.to_csv(OUTPUT / 'coefficient_diagnostics.csv')
display(diag.sort_values('t_stat', key=lambda s: s.abs(), ascending=False).head(10))
```
