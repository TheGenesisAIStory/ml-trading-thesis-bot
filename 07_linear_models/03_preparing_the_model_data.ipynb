# Preparing Model Data

Modernized for Python 3.10+ without legacy optional dependencies.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from linear_model_utils import regression_data, classification_data, panel_factor_data, ols_numpy, rank_ic
OUTPUT = FOLDER.parent / 'data' / 'linear_models'
OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
panel = panel_factor_data()
factors = [c for c in panel.columns if c.startswith('factor_')]
for col in factors:
    panel[f'{col}_rank'] = panel.groupby('date')[col].rank(pct=True)
panel['target_1d'] = panel.groupby('ticker')['return_1d'].shift(-1)
model_data = panel.dropna().copy()
model_data.to_parquet(OUTPUT / 'linear_model_panel.parquet', index=False)
display(model_data.head())
display(model_data.describe().T.head(12))
```
