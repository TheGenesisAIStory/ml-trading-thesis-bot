```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from linear_model_utils import regression_data, classification_data, panel_factor_data, ols_numpy, rank_ic
OUTPUT = FOLDER.parent / 'data' / 'linear_models'
OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
panel = panel_factor_data()
factor = 'factor_00'
ic = rank_ic(panel, factor)
panel['quantile'] = panel.groupby('date')[factor].transform(lambda s: pd.qcut(s.rank(method='first'), 5, labels=False) + 1)
quantile_returns = panel.groupby('quantile')['return_1d'].agg(['mean','std','count'])
ic.to_csv(OUTPUT / 'signal_rank_ic.csv', index=False)
quantile_returns.to_csv(OUTPUT / 'signal_quantile_returns.csv')
display(ic['ic'].describe().to_frame())
display(quantile_returns)
```
