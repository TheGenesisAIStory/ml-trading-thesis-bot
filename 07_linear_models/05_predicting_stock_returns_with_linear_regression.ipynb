# Predicting Returns with Linear Regression

Modernized for Python 3.10+ without legacy optional dependencies.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from linear_model_utils import regression_data, classification_data, panel_factor_data, ols_numpy, rank_ic
OUTPUT = FOLDER.parent / 'data' / 'linear_models'
OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

x, y = regression_data(seed=9)
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
models = {'ridge': RidgeCV(alphas=np.logspace(-3, 3, 13)), 'lasso': LassoCV(alphas=np.logspace(-3, 1, 9), max_iter=5000, random_state=42)}
rows=[]
for name, model in models.items():
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    rows.append({'model': name, 'r2': r2_score(y_test, pred), 'rmse': float(np.sqrt(mean_squared_error(y_test, pred)))})
summary = pd.DataFrame(rows)
summary.to_csv(OUTPUT / 'linear_prediction_summary.csv', index=False)
display(summary)
```
