# ARIMA-style AR Models

Modernized local version without fragile time-series dependencies.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from time_series_utils import make_time_series, make_pair_series, stationarity_summary, fit_ar, ewma_volatility, fit_var_two, hedge_ratio, pairs_backtest
OUTPUT = FOLDER.parent / 'data' / 'time_series_models'
OUTPUT.mkdir(parents=True, exist_ok=True)
ts = make_time_series()
pair = make_pair_series()
```

```python
coef, fitted = fit_ar(ts['stationary_ar'], lags=3)
forecast_eval = pd.DataFrame({'actual': ts.loc[fitted.index, 'stationary_ar'], 'fitted': fitted})
forecast_eval['error'] = forecast_eval['actual'] - forecast_eval['fitted']
coef.to_csv(OUTPUT / 'ar_coefficients.csv')
forecast_eval.to_parquet(OUTPUT / 'ar_forecast_eval.parquet')
display(coef.to_frame())
display(forecast_eval['error'].describe().to_frame())
```
