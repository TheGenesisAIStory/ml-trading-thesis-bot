# Pairs Trading Backtest

Modernized local version without fragile time-series dependencies.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from time_series_utils import make_time_series, make_pair_series, stationarity_summary, fit_ar, ewma_volatility, fit_var_two, hedge_ratio, pairs_backtest
OUTPUT = FOLDER.parent / 'data' / 'time_series_models'
OUTPUT.mkdir(parents=True, exist_ok=True)
ts = make_time_series()
pair = make_pair_series()
```

```python
bt = pairs_backtest(pair)
bt['cumulative_pnl'] = bt['strategy_return'].cumsum()
summary = pd.DataFrame([{'total_pnl': bt['strategy_return'].sum(), 'daily_pnl_std': bt['strategy_return'].std(), 'hit_rate': (bt['strategy_return'] > 0).mean(), 'max_drawdown': (bt['cumulative_pnl'] - bt['cumulative_pnl'].cummax()).min()}])
bt.to_parquet(OUTPUT / 'pairs_backtest.parquet', index=False)
summary.to_csv(OUTPUT / 'pairs_backtest_summary.csv', index=False)
display(summary)
display(bt.tail())
```
