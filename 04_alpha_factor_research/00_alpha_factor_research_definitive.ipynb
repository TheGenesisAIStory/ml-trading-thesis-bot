# Alpha Factor Research Definitive

Builds a reusable alpha factor panel with local fallbacks.

```python
from pathlib import Path
import sys
import pandas as pd

REPO_ROOT = Path.cwd().resolve().parent if Path.cwd().name == '04_alpha_factor_research' else Path.cwd().resolve()
if str(REPO_ROOT / '04_alpha_factor_research') not in sys.path:
    sys.path.insert(0, str(REPO_ROOT / '04_alpha_factor_research'))
from alpha_factor_utils import load_market_panel, compute_technical_factors, rank_ic, save_alpha_panel
OUTPUT = REPO_ROOT / 'data' / 'alpha_factor_research'
prices = load_market_panel(REPO_ROOT)
panel = compute_technical_factors(prices)
factor_cols = [c for c in panel.columns if c.endswith('_rank')]
ic = rank_ic(panel, factor_cols, target='return_5d')
save_alpha_panel(panel, OUTPUT / 'alpha_factor_panel.parquet')
ic.to_parquet(OUTPUT / 'rank_ic_5d.parquet', index=False)
display(panel.tail())
display(ic.groupby('factor')['ic'].agg(['mean','std','count']).sort_values('mean', ascending=False))
```
