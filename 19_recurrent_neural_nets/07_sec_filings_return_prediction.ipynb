# SEC Filing Return Prediction Fallback

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from rnn_utils import *

sns.set_theme(style='whitegrid', context='notebook')
pd.set_option('display.float_format', lambda x: f'{x:,.4f}')
OUTPUT = ensure_output_dir()
```

```python
data = text_corpus().assign(forward_return=[0.02, -0.03, 0.015, -0.02, 0.01, -0.015])
model = make_pipeline(TfidfVectorizer(), Ridge(alpha=1.0))
model.fit(data.text, data.forward_return)
pred = model.predict(data.text)
result = data.assign(prediction=pred)
result.to_csv(OUTPUT / 'sec_text_return_predictions.csv', index=False)
result
```
