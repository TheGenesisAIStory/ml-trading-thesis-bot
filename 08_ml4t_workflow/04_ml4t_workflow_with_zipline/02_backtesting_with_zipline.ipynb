# Backtesting without Zipline Runtime

Modernized workflow without pandas_datareader, Backtrader, or Zipline runtime dependencies.

```python
from pathlib import Path
import sys
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 80)
ROOT = Path.cwd().resolve()
if ROOT.name == '04_ml4t_workflow_with_zipline':
    WORKFLOW_DIR = ROOT.parent
else:
    WORKFLOW_DIR = ROOT
if str(WORKFLOW_DIR) not in sys.path:
    sys.path.insert(0, str(WORKFLOW_DIR))
from ml4t_workflow_utils import make_workflow_panel, vectorized_signal_backtest, walk_forward_ml, prediction_backtest
OUTPUT = WORKFLOW_DIR.parent / 'data' / 'ml4t_workflow'
OUTPUT.mkdir(parents=True, exist_ok=True)
panel = make_workflow_panel()
```

```python
returns = vectorized_signal_backtest(panel)
returns['benchmark'] = panel.groupby('date')['return_1d'].mean().reindex(returns['date']).to_numpy()
returns['active_return'] = returns['strategy_return'] - returns['benchmark']
returns.to_parquet(OUTPUT / 'zipline_style_backtest.parquet', index=False)
display(returns.tail())
```
