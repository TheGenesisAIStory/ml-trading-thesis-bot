# Private Active API Services

Local-only smoke checks for active API credentials. Outputs are masked and written to ignored `data/private_api_checks/`.

```python
from pathlib import Path
import sys
import pandas as pd

pd.set_option('display.max_columns', 80)
REPO_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'data' else Path.cwd().resolve()
PROVIDER_DIR = REPO_ROOT / '02_market_and_fundamental_data' / '03_data_providers'
if str(PROVIDER_DIR) not in sys.path:
    sys.path.insert(0, str(PROVIDER_DIR))
from api_provider_registry import load_dotenv, credential_status, configure_openbb_from_env, active_api_smoke_checks
PRIVATE_DIR = REPO_ROOT / 'data' / 'private_api_checks'
PRIVATE_DIR.mkdir(parents=True, exist_ok=True)
load_summary = {'.env.local': len(load_dotenv(REPO_ROOT / '.env.local')), '.env': len(load_dotenv(REPO_ROOT / '.env'))}
load_summary
```

```python
status = credential_status(mask=True)
status.to_csv(PRIVATE_DIR / 'credential_status_masked.csv', index=False)
openbb_configured = configure_openbb_from_env()
checks = active_api_smoke_checks(timeout=15)
checks.to_csv(PRIVATE_DIR / 'active_api_smoke_checks.csv', index=False)
checks.to_parquet(PRIVATE_DIR / 'active_api_smoke_checks.parquet', index=False)
display(status)
display(pd.DataFrame([{'openbb_credential': k, 'source_env_var': v} for k, v in sorted(openbb_configured.items())]))
display(checks[['name','configured','ok','status_code','detail']])
```
