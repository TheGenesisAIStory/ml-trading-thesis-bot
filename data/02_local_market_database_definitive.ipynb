# Local Market Database

Build a git-ignored SQLite base for market analyses. The notebook runs a small smoke database; use the CLI full command for the complete universe.

````python
from pathlib import Path
import sqlite3
import pandas as pd

from market_data_database import build_database, DB_PATH
````

````python
counts = build_database(DB_PATH, mode='smoke', years=5, max_symbols=8, allow_demo_fallback=True)
counts
````

````python
conn = sqlite3.connect(DB_PATH)
summary = pd.read_sql_query('SELECT asset_class, universe, COUNT(*) AS instruments FROM instruments GROUP BY asset_class, universe ORDER BY asset_class, universe', conn)
price_count = pd.read_sql_query('SELECT source, COUNT(*) AS rows FROM prices_daily GROUP BY source', conn)
conn.close()
summary, price_count
````
