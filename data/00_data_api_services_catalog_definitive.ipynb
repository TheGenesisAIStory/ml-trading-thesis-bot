# Data/API Services Catalog Definitive

Safe catalog for provider metadata, masked credential status, health checks, and local dataset inventory.

```python
from pathlib import Path
import sys
import pandas as pd

pd.set_option('display.max_columns', 80)
REPO_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'data' else Path.cwd().resolve()
PROVIDER_DIR = REPO_ROOT / '02_market_and_fundamental_data' / '03_data_providers'
if str(PROVIDER_DIR) not in sys.path:
    sys.path.insert(0, str(PROVIDER_DIR))
if str(REPO_ROOT / 'data') not in sys.path:
    sys.path.insert(0, str(REPO_ROOT / 'data'))
from api_provider_registry import providers_to_dataframe, credential_status, simple_health_checks, load_dotenv
from data_catalog import domains_to_dataframe, build_data_inventory, export_catalog
OUTPUT = REPO_ROOT / 'data' / 'catalog'
OUTPUT.mkdir(parents=True, exist_ok=True)
REPO_ROOT
```

```python
load_dotenv(REPO_ROOT / '.env.local')
load_dotenv(REPO_ROOT / '.env')
providers = providers_to_dataframe()
status = credential_status(mask=True)
health = simple_health_checks(timeout=8)
domains = domains_to_dataframe()
inventory = build_data_inventory(REPO_ROOT)
providers.to_csv(OUTPUT / 'api_providers.csv', index=False)
providers.to_parquet(OUTPUT / 'api_providers.parquet', index=False)
status.to_csv(OUTPUT / 'credential_status_masked.csv', index=False)
health.to_csv(OUTPUT / 'api_provider_health_checks.csv', index=False)
exports = export_catalog(REPO_ROOT, OUTPUT)
print(exports)
display(providers[['name','env_var','category','quality','free_tier']])
display(status)
display(health[['name','reachable','status_code']])
display(inventory[['domain','exists','files','size_mb','publish_to_git']])
```

Outputs are written under `data/catalog/`. They contain metadata and masked credential status only.
