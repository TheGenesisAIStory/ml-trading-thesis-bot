# Bias Variance Trade-Off

Modernized for Python 3.10+ with scikit-learn-only dependencies.

```python
from pathlib import Path
import sys
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from ml_process_utils import make_regression_frame, make_classification_frame, learning_curve_frame, validation_curve_frame
OUTPUT = FOLDER.parent / 'data' / 'machine_learning_process'
OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import Ridge

x, y = make_regression_frame(n_features=4, seed=17)
model = Pipeline([('poly', PolynomialFeatures(include_bias=False)), ('scale', StandardScaler()), ('ridge', Ridge(alpha=1.0))])
validation = validation_curve_frame(model, x, y, 'poly__degree', [1, 2, 3, 4], scoring='r2')
learning = learning_curve_frame(Pipeline([('scale', StandardScaler()), ('ridge', Ridge(alpha=1.0))]), x, y, scoring='r2')
validation.to_csv(OUTPUT / 'validation_curve.csv', index=False)
learning.to_csv(OUTPUT / 'learning_curve.csv', index=False)
display(validation)
display(learning)
```
