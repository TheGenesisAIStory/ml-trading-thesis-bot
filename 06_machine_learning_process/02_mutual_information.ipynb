# Mutual Information

Modernized for Python 3.10+ with scikit-learn-only dependencies.

```python
from pathlib import Path
import sys
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from ml_process_utils import make_regression_frame, make_classification_frame, learning_curve_frame, validation_curve_frame
OUTPUT = FOLDER.parent / 'data' / 'machine_learning_process'
OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif

x_reg, y_reg = make_regression_frame(seed=11)
x_cls, y_cls = make_classification_frame(seed=13)
reg_mi = pd.Series(mutual_info_regression(x_reg, y_reg, random_state=42), index=x_reg.columns, name='mutual_info_regression')
cls_mi = pd.Series(mutual_info_classif(x_cls, y_cls, random_state=42), index=x_cls.columns, name='mutual_info_classification')
summary = pd.concat([reg_mi.sort_values(ascending=False).head(8), cls_mi.sort_values(ascending=False).head(8)], axis=1)
summary.to_csv(OUTPUT / 'mutual_information_summary.csv')
display(summary)
```
