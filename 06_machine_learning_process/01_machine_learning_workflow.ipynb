# Machine Learning Workflow

Modernized for Python 3.10+ with scikit-learn-only dependencies.

```python
from pathlib import Path
import sys
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 80)
FOLDER = Path.cwd().resolve()
if str(FOLDER) not in sys.path:
    sys.path.insert(0, str(FOLDER))
from ml_process_utils import make_regression_frame, make_classification_frame, learning_curve_frame, validation_curve_frame
OUTPUT = FOLDER.parent / 'data' / 'machine_learning_process'
OUTPUT.mkdir(parents=True, exist_ok=True)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

x, y = make_regression_frame()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
pipe = Pipeline([('scale', StandardScaler()), ('model', KNeighborsRegressor(n_neighbors=8))])
cv_scores = cross_val_score(pipe, x_train, y_train, cv=5, scoring='r2')
pipe.fit(x_train, y_train)
pred = pipe.predict(x_test)
summary = pd.DataFrame([{'cv_r2_mean': cv_scores.mean(), 'cv_r2_std': cv_scores.std(), 'test_r2': r2_score(y_test, pred), 'test_rmse': float(np.sqrt(mean_squared_error(y_test, pred)))}])
summary.to_csv(OUTPUT / 'ml_workflow_summary.csv', index=False)
display(summary)
```
