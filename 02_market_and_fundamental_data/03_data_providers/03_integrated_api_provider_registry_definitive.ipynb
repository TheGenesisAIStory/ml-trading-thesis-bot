# Integrated API Provider Registry Definitive

Executable provider registry for the modern data/API layer.

```python
from pathlib import Path
import sys
import pandas as pd

pd.set_option('display.max_columns', 80)
REPO_ROOT = Path.cwd().resolve().parents[1] if Path.cwd().name == '03_data_providers' else Path.cwd().resolve()
if str(Path.cwd().resolve()) not in sys.path:
    sys.path.insert(0, str(Path.cwd().resolve()))
from api_provider_registry import providers_to_dataframe, credential_status, simple_health_checks, load_dotenv, configure_openbb_from_env
OUTPUT = REPO_ROOT / 'data' / 'data_providers' / 'api_registry'
OUTPUT.mkdir(parents=True, exist_ok=True)
load_dotenv(REPO_ROOT / '.env.local')
load_dotenv(REPO_ROOT / '.env')
providers = providers_to_dataframe()
status = credential_status(mask=True)
health = simple_health_checks(timeout=8)
providers.to_csv(OUTPUT / 'api_providers.csv', index=False)
providers.to_parquet(OUTPUT / 'api_providers.parquet', index=False)
status.to_csv(OUTPUT / 'credential_status_masked.csv', index=False)
health.to_csv(OUTPUT / 'api_provider_health_checks.csv', index=False)
print('OpenBB configured:', configure_openbb_from_env())
display(providers[['name','category','env_var','quality','free_tier']])
display(status)
display(health[['name','reachable','status_code']])
```

Credentials are never stored in this notebook. Use environment variables or ignored `.env.local`.
