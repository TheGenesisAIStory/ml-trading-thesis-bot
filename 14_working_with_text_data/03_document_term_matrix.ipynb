# Document-Term Matrix

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from text_data_utils import *

sns.set_theme(style='whitegrid', context='notebook')
pd.set_option('display.float_format', lambda x: f'{x:,.4f}')
OUTPUT = ensure_output_dir()
```

```python
corpus = sample_financial_corpus()
dtm = document_term_matrix(corpus.text)
tfidf = document_term_matrix(corpus.text, tfidf=True)
dtm.to_csv(OUTPUT / 'document_term_matrix.csv', index=False)
tfidf.to_csv(OUTPUT / 'tfidf_matrix.csv', index=False)
dtm.head()
```

```python
top_terms = dtm.sum().sort_values(ascending=False).head(15)
top_terms.plot(kind='bar', figsize=(9, 4), title='Top corpus terms')
plt.tight_layout()
```
