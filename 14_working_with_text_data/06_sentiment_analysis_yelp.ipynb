# Multiclass Review-Style Sentiment

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from text_data_utils import *

sns.set_theme(style='whitegrid', context='notebook')
pd.set_option('display.float_format', lambda x: f'{x:,.4f}')
OUTPUT = ensure_output_dir()
```

```python
reviews = pd.DataFrame({
    'rating': [5, 1, 4, 2, 5, 1],
    'text': [
        'excellent service strong execution and positive surprise',
        'poor delivery weak quality and negative experience',
        'good value and improved process',
        'slow response and disappointing result',
        'record performance and excellent support',
        'bad risk controls and severe losses',
    ]
})
reviews['sentiment'] = reviews.rating.map(lambda r: 'positive' if r >= 4 else 'negative')
model, results, accuracy = train_sentiment_classifier(reviews[['sentiment', 'text']])
results.to_csv(OUTPUT / 'review_sentiment_results.csv', index=False)
pd.Series({'accuracy': accuracy, 'rows': len(results)}).to_frame('value')
```
