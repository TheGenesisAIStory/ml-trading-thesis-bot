Safe, metadata-first catalog for free finance ML and alternative data sources.

```python
from pathlib import Path
import sys
import pandas as pd

REPO_ROOT = Path.cwd().resolve().parent if Path.cwd().name == '03_alternative_data' else Path.cwd().resolve()
ALT_DIR = REPO_ROOT / '03_alternative_data'
if str(ALT_DIR) not in sys.path:
    sys.path.insert(0, str(ALT_DIR))
from free_finance_ml_sources import sources_to_dataframe
OUTPUT = REPO_ROOT / 'data' / 'alternative_data' / 'free_finance_ml_sources'
OUTPUT.mkdir(parents=True, exist_ok=True)
sources = sources_to_dataframe()
sources.to_csv(OUTPUT / 'free_finance_ml_sources.csv', index=False)
sources.to_parquet(OUTPUT / 'free_finance_ml_sources.parquet', index=False)
display(sources)
print(OUTPUT)
```
